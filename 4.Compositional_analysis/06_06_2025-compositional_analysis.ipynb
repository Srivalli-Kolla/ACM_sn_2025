# Cell Composition - Overall

```python
cell_proportions_overall = overall.obs['celltype_de_group'].value_counts(normalize=True)
cell_proportions_overall
```

```python
composition_by_condition_overall = (
    overall.obs.groupby(['Condition', 'celltype_de_group'])
    .size()
    .unstack(fill_value=0)
)

composition_by_condition_prop_overall = composition_by_condition_overall.div(composition_by_condition_overall.sum(axis=1), axis=0)
```

```python
ax = composition_by_condition_prop_overall.T.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

plt.ylabel('Proportion of Cells')
plt.title('Cell Composition by Condition')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.xticks(rotation=45)

# Annotate each segment 
for i, patch_group in enumerate(ax.containers):
    for rect in patch_group:
        height = rect.get_height()
        if height > 0.05:  
            x = rect.get_x() + rect.get_width() / 2
            y = rect.get_y() + height / 2
            ax.text(
                x, y,
                f"{height:.2f}",
                ha='center', va='center',
                fontsize=8, color='white'
            )

plt.show()
```

# Cell Composition - ctr_noninf

```python
cell_proportions_ctr_noninf = ctr_noninf.obs['celltype_de_group'].value_counts(normalize=True)

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=cell_proportions_ctr_noninf.values,
    y=cell_proportions_ctr_noninf.index,
    palette='tab20'
)
plt.xlabel('Proportion of Cells')
plt.ylabel('Cell Type')
plt.title('Cell Composition - ctr_noninf')

# Add annotations
for p in ax.patches:
    width = p.get_width()
    y = p.get_y() + p.get_height() / 2
    ax.text(width + 0.01, y, f'{width:.2f}', va='center', fontsize=9)

plt.xlim(0, max(cell_proportions_ctr_noninf.values) + 0.1) 
plt.tight_layout()
plt.show()
```
